```python
all_data = [ca,us,af]
```

```python
df = pd.concat(all_data)
```

```python
df.posted_date = pd.to_datetime(df['posted_date'])
```

```python
df =  df.rename(columns  = {'onsite_remote':'mode'})
df.head()
```

```python
plt.figure(figsize=(40,15))
plt.style.use('seaborn-whitegrid')
ax=sns.countplot(x="location",data=df)
plt.xlabel('company')
plt.ylabel('Count')
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.25, p.get_height()+1), va = 'bottom', color = 'black')
```

```python
sns.countplot(ca.onsite_remote)
# Add labels and a title
plt.xlabel('Job Types')
plt.ylabel('Posted Jobs')
plt.title('Job Posts vs Job Types in Canada')
# Show the legend
plt.legend()

# Display the bar chart
plt.show()
```

```python
sns.countplot(us.onsite_remote)
# Add labels and a title
plt.xlabel('Job Types')
plt.ylabel('Posted Jobs')
plt.title('Job Posts vs Job Types in US')
# Show the legend
plt.legend()

# Display the bar chart
plt.show()
```

```python
sns.countplot(af.onsite_remote)
# Add labels and a title
plt.xlabel('Job Types')
plt.ylabel('Posted Jobs')
plt.title('Job Posts vs Job Types in Africa')
# Show the legend
plt.legend()

# Display the bar chart
plt.show()
```

```python
plt.figure(figsize = (15, 7))
company = df["company"].value_counts()[:30]
sns.barplot(x =company, y = company.index, palette = "Accent")
plt.xlabel("Count")
```

```python
plt.figure(figsize = (15, 7))
salary = df["salary"].value_counts()[:10]
sns.barplot(x =salary, y = salary.index, palette = "Accent")
plt.xlabel("Count")
```
